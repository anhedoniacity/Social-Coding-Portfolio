The Slave Voyages database is a collaborative project of multiple researchers and universities that compiles information about the majority of voyages that took place during the Trans-Atlantic Slave Trade. It compiles 1,651 data sources (mostly primary documentary sources, published secondary sources, and contemporary newspapers) from the start of the Trade up until the present. It appears that most of the voyage data included was originally collected as a matter of standard economic and government recordkeeping, without any thought to the moral implications; as a result, the database is substantially lacking in information about enslaved people themselves, who were seen as product. However, the database does include some data about captive Africans collected by naval operations that sought to suppress the slave trade in the 1800s. Overall it is not a "clean" dataset, and contains a substantial amount of imputed data, but it is a nearly complete record of all voyages that occurred in the Trans-Atlantic Slave Trade.
